```python
import pandas as pd
```

```python
df = pd.read_csv("aggregate_travel_time.csv")
df2 = pd.read_csv("median_income.csv")
df3 = pd.read_csv("percent_bachelor_or_higher.csv")
```

```python
df = df.drop(columns = ["GEO.id","GEO.id2"])
df2 = df2.drop(columns = ["GEO.id","GEO.id2"])
df3 = df3.drop(columns = ["GEO.id","GEO.id2"])
```

```python
df_main0 = pd.merge(df,df2, on = "GEO.display-label")
df_main = pd.merge(df_main0, df3, on = "GEO.display-label")
```

```python
headers = df_main.iloc[0]
df_main_new = pd.DataFrame(df_main.values[1:], columns = headers)

df_main_new_renamed = df_main_new.rename(columns={'Geography': "city",'Estimate; Aggregate travel time to work (in minutes):':"agg_commute",'Median income (dollars); Estimate; Households':"house_income_median", "Percent; Estimate; Percent bachelor's degree or higher" :"bachelors_or_higher_percent"})

city_df = df_main_new_renamed

city_df.head()
```

```python
new_city_df = city_df["city"].str.split(", ", n =1, expand= True )
```

```python
city_df["city"] = new_city_df[0]
new_city_df.head()
```

```python
city_df["state"] = new_city_df[1]
city_df.head()
```

```python
reorder = ["city","state","agg_commute","house_income_median","bachelors_or_higher_percent"]
```

```python
city_df = city_df.reindex(columns=reorder)
```
